## Inspecting Columns
For each column, print the number of unique values and the number of missing (null) values.


```python
for col in Data_jobs.columns:
    print(f"Column {col}:")
    print(f"Number of unique values: {len(Data_jobs[col].value_counts())}")
    print(f"Number of null values: {Data_jobs[col].isnull().sum()}")
    print("------------------------------")
```

## Checking for Duplicates
